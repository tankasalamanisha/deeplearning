## Linear Regression implimentation using traditional ML vs Deep Learning.

````python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Using TensorFlow/Keras
import tensorflow as tf
import numpy as np
````

````python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2,4,5,4,5])

#Sklearn implimentation
model = LinearRegression()
model.fit(X, y)

#Predictions
preds=model.predict(X)
print("Sklearn implimentation Coefficients:",model.coef_)

X = np.array([[1], [2], [3], [4], [5]], dtype=float)
y = np.array([2, 4, 5, 4, 5], dtype=float)

# Build neural network (single layer = linear regression)
# Sequential --> means the data flows linearly across layers : input --> layer1 --> layer2 --> output
# Dense --> means a fully connected network where each input is connected to every neuron in the layer. 
# units --> Number of predictions/outputs in this layer. As we are doing linear regression/prediction, we only need one neuron in the layer. Thus, the unit = 1. If we are predicting multiple outputs, we might need Dense(units=D), where D --> number of outputs.
# input_shape --> Shape of the input data. Here, we have only one feature/input, The comma makes it a tuple (Python requirement), Only needed for the **first layer** so Keras knows what input to expect If you had 3 features (e.g., bedrooms, bathrooms, square feet), you'd use `input_shape=(3,)`
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=(1,))
])

model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(X, y, epochs=100, verbose=True)

dl_preds = model.predict(X)
print("Deep Learning implimentation Weights:", model.layers[0].get_weights())
````
